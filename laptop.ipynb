```
import pandas as pd
from google.colab import files
uploaded =files.upload()
```

```
import io
data = pd.read_csv(io.BytesIO(uploaded['Laptops.csv']),encoding="ISO-8859-1")
change=data
change
```

```
change[['ScreenResolution']]=data[['ScreenResolution']].fillna(method='pad')   #changing categorical values to precious rows
change
```

```
m=data['Price_euros'].mean()  #finding the average of the prices colums
change[['Price_euros']]=data[['Price_euros']].fillna(m)   #changing numerical NaN values to average values of the colum
change
```

```
change['Inches']=data['Inches'].interpolate()      #interpolation of the inches column
change
```

```
from matplotlib import pyplot
# histogram plot
pyplot.hist(change['Price_euros'])
pyplot.show()
```

```
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot
qqplot(change['Price_euros'], line='s')
pyplot.show()                                       #Right skewed(concave up) so not normal
```

```
from matplotlib import pyplot
# histogram plot
pyplot.hist(change['Weight'])
pyplot.show()
```

```
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot
qqplot(change['Weight'], line='s')
pyplot.show()                                         #Checking normality for Weight of laptops
```

```
# Shapiro-Wilk Test
from scipy.stats import shapiro
# normality test
stat, p = shapiro(change['Inches'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Normal (fail to reject H0)')
else:
	print('Sample does not look Normal (reject H0)')
```
